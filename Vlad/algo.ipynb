```python
import pandas

excel_data_df = pandas.read_excel('Distance20221022.xlsx', sheet_name='Roads')


graph = {}

for i, row in excel_data_df.iterrows():
  
  # print(f"Index: {i}")
  # print(type(row['source_point_id']))
  if row['source_point_id'] in graph:
    graph[int(row['source_point_id'])][int(row['target_point_id'])] = row['time']*60
  else:
    graph[int(row['source_point_id'])] = {int(row['target_point_id']): row['time']*60} 

graph
```

```python
g=graph
d = {}
mark = {}
for i,item in graph.items():
    mark[i]=False
    d[i]=10000
    
    for j in item:
      mark[j]=False
      d[j]=10000
d[1109] = 0
for i in graph:
    v = -1
    for j in graph:
        if (not mark[j]) and ((v == -1) or (d[v] > d[j])):
            v = j
    mark[v] = True
    for k in graph[v]:
      if d[k] > d[v] + g[v][k]:
        d[k] = d[v] + g[v][k]
print(d)

ssd = sorted(d, key=d.get)
k = ssd[1]
print(ssd)
{k: d[k]}
```
